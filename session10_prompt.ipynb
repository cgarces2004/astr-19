```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu, sigma = 0, 1
random_numbers_normal = np.random.normal(mu, sigma, 1000)
```

```python
plt.hist(random_numbers_normal, bins=100)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Random Numbers from a Normal Distribution')

plt.savefig('normal_distribution_histogram.pdf')
```
