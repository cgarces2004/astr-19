```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def exp_function(x):
    return np.exp(x)
```

```python
y = exp_function(x)

plt.figure()
plt.plot(x, y)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
plt.title("Exponential Growth of Awesomeness Over Time")

plt.savefig("awesomeness_over_time.pdf")
```
